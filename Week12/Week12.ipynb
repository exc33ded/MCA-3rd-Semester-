### 1. Write a shell script to back up a directory to a specified location.

1. First create an empty file with `.sh` extension, by writing `nano backup.sh`.
2. Enter the commands:<br>

`source_dir=$1`<br>
`destination_dir=$2`<br>

`if [ -d "$source_dir" ]; then`<br>
`  if [ ! -d "$destination_dir" ]; then`<br>
`    mkdir -p "$destination_dir"`<br>
`  fi`
`  cp -r "$source_dir"/* "$destination_dir"`<br>
`  echo "Backup of $source_dir completed successfully to $destination_dir."`<br>
`else`<br>
`  echo "Source dir $source_dir does not exist."`<br>
`fi`<br>


3. Save by `CTRL + X`, then press `Y` and then `Enter`
4. Provide excecuting permission by `chmod +x backup.sh`
5. Run the file by typing ` ./backup.sh /root/week12/folder /root/week12/backupFolder`

### 2. Write a shell script to monitor disk usage and send an alert if usage exceeds a threshold.

1. First create an empty file with `.sh` extension, by writing `nano monitor_disk_usage.sh`.
2. Enter the commands:<br>

`threshold=$1`<br>

`current_usage=$(df / | grep / | awk '{ print $5 }' | sed 's/%//g')`<br>

`if [ "$current_usage" -gt "$threshold" ]; then`<br>
`  echo "Alert: Disk usage is above $threshold%. Current usage is $current_usage%."`<br>
`else`<br>
`  echo "Disk usage is under control. Current usage is $current_usage%."`<br>
`fi`<br>

3. Save by `CTRL + X`, then press `Y` and then `Enter`
4. Provide excecuting permission by `chmod +x monitor_disk_usage.sh`
5. Run the file by typing `./monitor_disk_usage.sh 80`

### 3. Write a shell script to automate the creation of user accounts.

1. First create an empty file with `.sh` extension, by writing `nano create_users.sh`.
2. Enter the commands:<br>

`username_list_file=$1`<br>

`while IFS= read -r username; do`<br>
`  if id "$username" &>/dev/null; then`<br>
`    echo "User $username already exists."`<br>
`  else`<br>
`    sudo useradd "$username"`<br>
`    echo "User $username created successfully."`<br>
`  fi`<br>

### Consider two features x and y based on the following function: y = x1^2 + 3x2 + c, where c can be prepared based on 1000 random values between 0 and 1 Now generate 1000 random values between 0 and 1 for x1 and x2. Calculate y based on above function. Now train Polynomial Regression model and check the score for the same.

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
x1 = np.random.rand(1000)
```

```python
x2 = np.random.rand(1000)
```

```python
c = np.random.rand(1000)
```

```python
y = x1**2 + 3*x2 + c
```

```python
X = np.column_stack((x1, x2))
```

```python
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)
```

```python
model = LinearRegression()
model.fit(X_poly, y)
```

```python
y_pred = model.predict(X_poly)
score = r2_score(y, y_pred)
print(f"R^2 Score: {score}")
```
